### Tratamento de dados 

###### Visualização para entender a presença e distribuição de dados ausentes

```python
msno.bar(df)
```

```python
df = df.dropna(subset = {"default"}) # exclusão das linhas nulas
```

```python
df.describe() #resumo das variaveis quantitativas. 
```

Baseado na descrição, o ideal é a substituiçao dos valores faltantes do limite de credito e do numero de emprestimos com inadiplencia pela mediana

```python
df.head(20)
```

```python
valores_subst = {'limite_credito': df['limite_credito'].median(),'n_emprestimos_inadiplentes':df['n_emprestimos_inadiplentes'].median()}
df = df.fillna(value=valores_subst) #substituindo os valores nos campos NaN pela mediana
df.head(20)
```

###### Substituição das variaveis qualitativas pela moda (maior frequencia)

```python
df['genero'].value_counts() #observa-se o valor de maior frequencia (masculino)
```

```python
df['perfil_facebook'].value_counts()#observa-se o valor de maior frequencia
```

```python
df['signo'].value_counts() #observa-se que não existe valor de maior frequencia
```

```python
#substituindo pela moda nos valores NAN:
df['genero'] =df['genero'].fillna(df['genero'].mode()[0])
df['perfil_facebook'] =df['perfil_facebook'].fillna(df['perfil_facebook'].mode()[0])
```

##### Transformando as variáveis com duas categorias em variáveis binárias para facilitar a modelagem (dummies)


```python
df=pd.get_dummies(df,columns=['prazo_pagamento','genero','perfil_facebook','default'],drop_first=True)
```

```python
df.head()
```

### Tratamento de outlier - Substituindo outlier por quantil:

```python
#visualização
sns.boxplot(data=df, x='default_True',y='t_risco',palette="pastel")
   
```

```python
out_max= df['t_risco'].quantile(0.975)
out_max
```

```python
x = df.loc[df['t_risco'] > out_max, 't_risco'] #substituição de todos os valores acima de 0.51 por 0.51
x
```

```python
df.loc[df['t_risco'] > out_max, 't_risco'] =out_max #substituição de todos os valores acima
```

```python
out_min= df['t_risco'].quantile(0.025)
out_min
```

```python
df.loc[df['t_risco'] < out_min, 't_risco'] = out_min #substituição de todos os valores acima de 0.51 por 0,51
```

```python
#novo grafico
sns.boxplot(data=df, x='default_True',y='t_risco',palette="pastel")
   
```
